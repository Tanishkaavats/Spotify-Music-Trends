```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()


```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# 1. Year-wise Song Count
songs_per_year = df['year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
sns.lineplot(x=songs_per_year.index, y=songs_per_year.values)
plt.title("Number of Songs Released per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Popularity vs Danceability
plt.figure(figsize=(8,6))
sns.scatterplot(data=df.sample(2000), x="danceability", y="popularity", alpha=0.5)
plt.title("Popularity vs Danceability")
plt.show()
```

```python
# 3. Popularity vs Energy
plt.figure(figsize=(8,6))
sns.scatterplot(data=df.sample(2000), x="energy", y="popularity", alpha=0.5)
plt.title("Popularity vs Energy")
plt.show()
```

```python
# 4. Average Danceability & Energy per Year
avg_features = df.groupby('year')[['danceability', 'energy']].mean()
avg_features.plot(figsize=(12,6))
plt.title("Danceability & Energy Trends Over Time")
plt.show()
```
